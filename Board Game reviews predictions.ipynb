```python
df_game=df_game[df_game['users_rated'] > 0]

df_game=df_game.dropna(axis=0)


plt.hist(df_game['average_rating'])
plt.show()
```

```python
columns=df_game.columns.tolist()


columns=[c for c in columns if c not in['bayes_average_rating','average_rating','type','name','Id']]

target='average_rating'
```

```python
df_game.shape
```

```python
from sklearn.model_selection import train_test_split


train=df_game.sample(frac=0.8,random_state=1)

test= df_game.loc[~df_game.index.isin(train.index)]


```

```python
print(train.shape)

print(test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
log=LinearRegression()
```

```python
log.fit(train[columns],train[target])
```

```python
predictions=log.predict(test[columns])
```

```python
from sklearn import metrics
```

```python
print('MSE:',metrics.mean_squared_error(predictions, test[target]))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RG=RandomForestRegressor()
```

```python
RG.fit(train[columns],train[target])
```

```python
predictions=RG.predict(test[columns])
```

```python
from sklearn import metrics
```

```python
print('MSE:',metrics.mean_squared_error(predictions, test[target]))
```
